```python
# read all matched trajectories generated before using FMM
import os
trajectories = []
with open('/content/trajectories_output.txt', 'r') as f:
    for line in f:
      # some lines are empty
      if line != '\n':
        traj_list = []
        for x in line.split(','):
          lat, lon = x.split(" ")
          traj_list.append((float(lat), float(lon)))

        trajectories.append(traj_list)
```

```python
# plot trajectories
import folium
from folium.plugins import HeatMap

map = folium.Map(zoom_start=10, height=700, location=[33.5186, -86.8104])
for d in trajectories[:100]:
  HeatMap(d, blur=5, radius=5).add_to(map)

map
```

```python
# decompose trajectories into points and plot
trajs_as_points = []
for r in trajectories:
  for rr in r:
    trajs_as_points.append(rr)

print(len(trajs_as_points))
```

```python
# plot first x trajectory points
x = 5000
map = folium.Map(zoom_start=10, height=700, location=[33.5186, -86.8104])
HeatMap(trajs_as_points[:x], blur=5, radius=5).add_to(map)
map
```

```python
# sample k points and plot
import random 
k = 5000
samples = random.choices(trajs_as_points, k=k)
# plot all traj together to see congestion
map = folium.Map(zoom_start=10, height=700, location=[33.5186, -86.8104])
HeatMap(samples, blur=5, radius=5).add_to(map)
map
```
